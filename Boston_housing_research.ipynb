```
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```
# Load boston housing datasets from sklearn and print it
from sklearn.datasets import load_boston
boston = load_boston ()
print(boston)
```

```
#transform the datset into a dat frame
#data - the date we want or the independent variable as the x values
#feature _ names of the data columns
#target variable for the price of the houses or dependent variable also known as y value
df_x = pd.DataFrame(boston.data, columns=boston.feature_names)
df_y = pd.DataFrame(boston.target)
```

```
#get some statistics from the data set
df_x.describe()
```

```
#Initialize linear regression model
reg = linear_model.LinearRegression()
```

```
#split data into 67% traiingiand 33% testing data
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size= 0.33, random_state=42)
```

```
#Train the model with our training data
reg.fit(x_train, y_train)
```

```
#print the coefficients/weights for each feature/ colum of our model  
print(reg.coef_) 
```

```
#print the predictions on our test data
y_pred = reg.predict(x_test)
print(y_pred)
```

```
#print the actual values
print(y_test)
```

```
#check the model performane accuracy using Mean Squared Error MSE
print(np.mean((y_pred - y_test)**2))
```

```
#check the model performane accuracy using Mean Squared Error MSE and sklearn.metrics
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, y_pred))
```
